```python
# Step 1: Put all your DataFrames in a list
dfs = [pilot, wave_2, wave_3, wave_4, wave_5, wave_6]

# Step 2: Find common columns across all DataFrames
common_cols = set(dfs[0].columns)
for df in dfs[1:]:
    common_cols &= set(df.columns)

# Step 3: Keep only *uniquely named* common columns for each DataFrame
def get_clean_df(df, common_cols):
    df = df.loc[:, ~df.columns.duplicated()]  # drop duplicate columns
    return df[list(common_cols)]

dfs_cleaned = [get_clean_df(df, common_cols) for df in dfs]

# Step 4: Concatenate the cleaned DataFrames
all_waves = pd.concat(dfs_cleaned, ignore_index=True)
```
