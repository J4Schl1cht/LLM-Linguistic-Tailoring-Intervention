```python
response_w6["Status"] = response_w6['Status'].replace({
    'Completed': 1,
    'Unknown': 99,
    'Expired': 0,
    'Canceled': 99,
    'Blocked': 99,
     'NaN': 99})

```

```python
# Ensure Status is numeric
response_w6["Status"] = response_w6["Status"].astype(int)

# Sort so that Status==1 comes last in case of tie-break
response_w6 = response_w6.sort_values(by=["User ID", "Activity ID", "Status"])

# Define the function to resolve duplicates
def resolve_duplicates(group):
    status_1_rows = group[group["Status"] == 1]
    if len(status_1_rows) > 0:
        return status_1_rows.iloc[[0]]  # Keep the FIRST row with Status == 1
    else:
        return group.iloc[[-1]]  # Keep last row if no Status == 1 found

# Apply it
deduped = (
    response_w6
    .groupby(["User ID", "Activity ID"], group_keys=False)
    .apply(resolve_duplicates)
    .reset_index(drop=True)
)
```

```python
# Pivot the dataframe
response_w6_wide = deduped.pivot(index='User ID', columns='Activity ID', values='Status')

response_w6_wide = response_w6_wide.reset_index()
response_w6_wide = response_w6_wide.merge(keys_w6[["ID", "Norstat ID", "condition"]], left_on="User ID", right_on="ID", how ="left")
```
